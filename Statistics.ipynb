Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
-  In hypothesis testing, decisions are made based on sample data, but there is always a chance of making incorrect conclusions. These mistakes are known as Type I and Type II errors.

1. Type I Error (False Positive)

Type I error occurs when we reject the null hypothesis (H₀) even though it is actually true.

It represents a false alarm.

Probability of Type I error = α (significance level), commonly 0.05.

Example:

A medical test says a person has a disease, but the person is actually healthy.

Impact:

Can lead to wrong claims, unnecessary actions, or financial loss.

In critical fields (medicine, aviation), controlling Type I error is extremely important.

2. Type II Error (False Negative)

Type II error occurs when we fail to reject the null hypothesis even though it is false.

It represents missing a real effect.

Probability of Type II error = β.

Power of a test = 1 – β.

Example:

A medical test fails to detect a disease that the person actually has.

Impact:

Leads to missed opportunities, undetected risks, or false sense of security.

-  Conclusion:

Type I and Type II errors directly affect the reliability of decisions. Choosing the right significance level (α) and ensuring sufficient sample size helps balance both errors and make better statistical decisions.

Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?


-  A P-value is the probability of observing sample results at least as extreme as the actual data if the null hypothesis (H₀) is true.

It measures the strength of evidence against the null hypothesis.

How to Interpret P-value:

P-value	Interpretation	Decision
P ≤ α (0.05)	Strong evidence against H₀

Reject H₀

P > α	Weak evidence against H₀

Fail to reject H₀


Key Points:

Small P-value → strong evidence against H₀.

Large P-value → data is consistent with H₀.

P-value is NOT

the probability that H₀ is true

the probability that the result is due to chance

It only tells how surprising the data is if H₀ were true.

Example:

If P = 0.03 (less than 0.05), then the data is unlikely under H₀ → we reject H₀.

Question 3: Explain the difference between a Z-test and a T-test, including when to use each.
Answer:

A Z-test and a T-test are both statistical tests used to compare sample data with population parameters, but they are used in different situations based on sample size and availability of population standard deviation.

1. Population Standard Deviation (σ):

Z-test is used when the population standard deviation is known.

T-test is used when the population standard deviation is unknown, and we must estimate it using the sample standard deviation.

2. Sample Size Condition:

Z-test is usually performed when the sample size is large, typically n > 30.

T-test is preferred when the sample size is small, usually n ≤ 30, because small samples introduce more uncertainty.

3. Distribution Used:

Z-test uses the normal distribution (standard normal distribution).

T-test uses the t-distribution, which has heavier tails to account for extra uncertainty when sample size is small.

4. When to Use Each Test:
Use the Z-test when:

You know the population standard deviation.

The sample size is large (Central Limit Theorem applies).

Data is approximately normally distributed.

Use the T-test when:

Population standard deviation is unknown.

Sample size is small (n ≤ 30).

Data comes from a normally distributed population.

-  Conclusion:

The Z-test is appropriate when data conditions are well-defined and sample size is large, while the T-test is more flexible and is used when dealing with small samples or unknown variance. The choice between them ensures accurate hypothesis testing.

Question 4:What is a confidence interval, and how does the margin of error influence
its width and interpretation?

-  A Confidence Interval (CI) is a range of values that likely contains the true population parameter (mean, proportion, etc.).

Example 95% CI:
(50 ± 3) means we are 95% confident that the true mean lies between 47 and 53.

Margin of Error (ME):

Formula:
ME = Critical Value × Standard Error

How ME affects CI width:

Larger ME → wider CI (less precision)

Smaller ME → narrower CI (more precision)

Factors that increase ME (wider CI):

Higher confidence level (99% > 95%)

Higher standard deviation

Smaller sample size

Conclusion:

Confidence intervals express uncertainty in estimates. Margin of error directly controls how wide or narrow the interval is.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?


-  Purpose of ANOVA (Analysis of Variance):

ANOVA tests whether three or more group means are statistically different.

It compares:

Between-group variance

Within-group variance

If between-group variance is significantly higher → means differ.

Assumptions of ANOVA:

Independence of observations

Normality within each group

Equal variances (homogeneity of variance)

How ANOVA extends hypothesis testing:

t-test compares only two means.

ANOVA compares 3 or more means at once and avoids multiple testing errors.

Hypotheses:

H₀: All means are equal

H₁: At least one mean is different

Conclusion:

ANOVA is a powerful method for multi-group comparisons and avoids errors caused by running many t-tests.

```python
# Question 6: Write a Python program to perform a one-sample Z-test and interpret the
# result for a given dataset.

import numpy as np
from statsmodels.stats.weightstats import ztest

data = [52, 49, 50, 51, 48, 53, 52, 50, 49, 51]

mu_0 = 50


z_stat, p_value = ztest(data, value=mu_0)

print("Z-statistic:", z_stat)
print("P-value:", p_value)


if p_value < 0.05:
    print("Reject H0: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject H0: No significant difference.")
```

```python
# Question 8: Generate multiple samples from a non-normal distribution and implement
# the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=1, size=100000)

sample_means = []


for i in range(1000):
    sample = np.random.choice(population, size=50)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT Demonstration (Means of Non-normal Distribution)")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()
```

```python
# Question 9: Write a Python function to calculate and visualize the confidence interval
# for a sample mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2, len(data) - 1)

    lower = mean - margin
    upper = mean + margin

    print(f"{confidence*100}% Confidence Interval: ({lower:.2f}, {upper:.2f})")


    plt.errorbar(1, mean, yerr=margin, fmt='o')
    plt.title("Confidence Interval for Sample Mean")
    plt.xlim(0, 2)
    plt.ylabel("Value")
    plt.show()

data = [5, 6, 7, 8, 9, 10]
plot_confidence_interval(data)
```

```python
# Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
# observed and expected distributions, and explain the outcome.
import numpy as np
from scipy.stats import chisquare


observed = np.array([40, 30, 20, 10])
expected = np.array([25, 25, 25, 25])


chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)


if p_value < 0.05:
    print("Reject H0: Observed distribution is significantly different from expected.")
else:
    print("Fail to reject H0: No significant difference.")
```
