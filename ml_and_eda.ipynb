Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.


-  AI (Artificial Intelligence): Goal of creating systems that perform tasks that normally require human intelligence.

- AI Scope: Very broad â€” includes reasoning, planning, learning, perception, language understanding.

- AI Techniques: Rule-based systems, search algorithms, knowledge representation, planning, logic.

- AI Applications: Robotics, game-playing agents, decision support systems, automated reasoning.

-  ML (Machine Learning): Subset of AI focused on algorithms that learn patterns from data.

- ML Scope: Model-building from data to make predictions or decisions without explicit programming for the task.

- ML Techniques: Supervised (regression, classification), unsupervised (clustering), reinforcement learning.

- ML Applications: Spam detection, recommendation systems, predictive maintenance, price forecasting.

- DL (Deep Learning): Subset of ML using multi-layer neural networks to learn hierarchical representations.

- DL Scope: Best for large datasets and complex inputs (images, audio, text).

- DL Techniques: Convolutional Neural Networks (CNN), Recurrent Neural Networks (RNN), Transformers.

- DL Applications: Image recognition, speech-to-text, language models, autonomous driving.

- Data Science: Interdisciplinary field combining data collection, cleaning, analysis, visualization, and modeling to extract insights.

- Data Science Scope: End-to-end: data engineering + statistics + ML + communication to stakeholders.

- Data Science Techniques: EDA, statistical testing, feature engineering, ML modeling, dashboards.

- Data Science Applications: Business analytics, A/B testing, forecasting, KPI dashboards.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?


- Overfitting â€” definition: Model fits noise and specific patterns of training data, not general patterns.

- Overfitting â€” symptom: High training accuracy, much lower validation/test accuracy.

- Overfitting â€” cause: Model too complex, too many parameters, not enough data.

- Underfitting â€” definition: Model too simple to capture underlying data patterns.

- Underfitting â€” symptom: Low training accuracy and low validation/test accuracy.

- Underfitting â€” cause: Model with insufficient complexity or poor features.

- Detecting (common): Compare training vs validation/test scores; plot learning curves.

- Detecting (cross-validation): Large variance across folds suggests overfitting; uniformly low scores suggest underfitting.

- Prevent overfitting â€” regularization: L1/L2 penalties, dropout (for neural nets).

- Prevent overfitting â€” simpler model: Reduce complexity (prune trees, reduce layers).

- Prevent overfitting â€” more data: Collect or augment data.

- Prevent overfitting â€” early stopping: Stop training when validation loss increases.

- Prevent underfitting â€” increase complexity: Add features, use a more powerful model.

- Prevent underfitting â€” feature engineering: Create informative features, remove noise.

- Use cross-validation to tune hyperparameters and balance biasâ€“variance tradeoff.

Question 3: How would you handle missing values in a dataset? Explain at least three methods with examples.


- Method 1 â€” Deletion (Listwise): Remove rows with missing values when missingness is rare (< ~5%).

- Deletion â€” example: If 2 of 10,000 rows have missing Age, drop those rows.

- Deletion â€” caveat: Can bias results if missingness is not random.

- Method 2 â€” Mean/Median Imputation: Replace missing numeric values with column mean or median.

- Mean/Median â€” example: Age column [22, 25, NaN, 30] â†’ median 25 â†’ replace NaN with 25.

- Mean/Median â€” caveat: Underestimates variance, may distort distributions.

- Method 3 â€” Predictive Imputation (Model-based): Train a model (e.g., RandomForest) to predict missing values from other features.

- Predictive â€” example: Use features (Pclass, Sex, Fare) to predict missing Age on Titanic.

- Predictive â€” benefit: Captures relationships, usually better than naive imputation.

- Method 4 â€” KNN Imputation: Use k-nearest rows to impute values based on similar samples.

- KNN â€” example: For missing income, average incomes of 5 nearest neighbors.

- KNN â€” caveat: Slower on large datasets, sensitive to scaling.

- Method 5 â€” Flag + Impute: Create a binary flag indicating missingness and then impute (mean/median).

- Flag â€” purpose: Model can learn if missingness itself is informative.

Question 4: What is an imbalanced dataset? Describe two techniques to handle it

- Imbalanced dataset â€” definition: One class (minority) has far fewer samples than another (majority).

- Imbalanced â€” problem: Standard models bias toward majority, poor minority recall/precision.

- Technique 1 â€” Random Over/Under Sampling (theory): Adjust class counts by duplicating minority (oversample) or removing majority (undersample).

- Over/Under â€” practical: Use imblearn.RandomOverSampler or RandomUnderSampler.

- Over/Under â€” pros/cons: Oversampling can cause overfitting; undersampling may lose information.

- Technique 2 â€” SMOTE (theory): Synthetically create minority examples by interpolating between minority neighbors.

- SMOTE â€” practical: Use imblearn.SMOTE().fit_resample(X, y).

- SMOTE â€” pros/cons: Reduces overfitting from duplication, but can create noisy borderline samples.

- Technique 3 â€” Class weights / cost-sensitive learning (theory): Penalize misclassification of minority class more heavily during training.

- Class weights â€” practical: Set class_weight='balanced' in scikit-learn classifiers or pass custom weights.

- Class weights â€” pros/cons: Works without changing data distribution; depends on model supporting weights.

- Evaluation note: Use precision/recall, F1, ROC-AUC, PR-AUC â€” accuracy is misleading for imbalanced data.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?


-  Label Encoding â€” what: Map categories to integer labels (e.g., Red=0, Green=1, Blue=2).

- Label Encoding â€” use-case: Ordinal variables where order matters (e.g., Low < Medium < High).

- Label Encoding â€” caveat: For nominal data it may introduce false ordinal relationship.

- One-Hot Encoding â€” what: Create binary column per category (Red â†’ [1,0,0]).

- One-Hot â€” use-case: Nominal categorical variables with no intrinsic order (e.g., color, country).

- One-Hot â€” caveat: High-cardinality variables create many columns (curse of dimensionality).

- When prefer Label: Ordinal categories or tree-based models where label integers may be okay.

- When prefer One-Hot: Linear models, distance-based models, or when no order exists among categories.



```python
#Question 7: Google Play Store Dataset â€” a) Relationship between app categories and ratings.

import pandas as pd
df = pd.read_csv("googleplaystore.csv")
category_ratings = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)
print(category_ratings.head(10))
print(category_ratings.tail(10))
```

- Line â€” what code does: Groups apps by Category and computes average Rating per category.

- Expected top categories (typical): Events, Education, Art & Design often show higher average ratings.

- Expected bottom categories (typical): Weather, Dating, Finance often show lower average ratings.

- Reasons â€” high rating categories: Smaller user base, niche helpful functionality, lower expectations, stable features.

- Reasons â€” low rating categories: Higher expectations, frequent bugs, monetization issues, device compatibility problems.

- EDA tips: Show boxplots per category, median + IQR, count per category for context.

Example visualization code:

```python
# Example visualization code:
import matplotlib.pyplot as plt
category_ratings.plot(kind='bar', figsize=(12,6))
plt.ylabel("Average Rating")
plt.title("Average App Rating by Category")
plt.show()
```

```python
# Question 8: Titanic Dataset â€” a) Survival by Pclass; b) Age effect (children vs adults).
import pandas as pd
df = pd.read_csv("titanic.csv")
pclass_survival = df.groupby("Pclass")["Survived"].mean()
df['Group'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
age_group_survival = df.groupby("Group")["Survived"].mean()
print(pclass_survival)
print(age_group_survival)
```

- Pclass result (typical): 1st class shows highest survival rate, 3rd class lowest.

- Why Pclass effect: 1st-class passengers had better access to lifeboats, cabins closer to deck, social priority.

- Age grouping result (typical): Children (Age < 18) have higher survival proportion than adults.

- Why children survived more: "Women and children first" evacuation policy, crew priority, social rescue behavior.

- Additional EDA: Cross-tab survival by Pclass and Sex to see combined effects (e.g., female first-class highest survival).

```python
# Question 9: Flight Price Prediction Dataset â€” a) Days left vs price; b) Airline comparison for same route.
# Code for days-left trend (short):
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("flight_price.csv")
mean_prices = df.groupby("days_left")["price"].mean()
mean_prices.plot()
plt.xlabel("Days left to departure")
plt.ylabel("Average price")
plt.title("Flight price vs days left")
plt.show()



```

-  Typical pattern: Prices relatively stable far from departure, then increase gradually and often surge exponentially within last 7â€“10 days.

- Best booking window (typical recommendation): Around 25â€“40 days before departure for domestic routes; varies by route and season.

- Typical results: Low-cost carriers (e.g., Indigo/SpiceJet/AirAsia depending on dataset) appear cheaper; full-service carriers (Vistara/Air India) are pricier.

- Reasons airline differences: Service level, baggage policy, demand, seat availability, and hub/slot costs.

```python
# Code to check correlations & visuals (short):
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("hr_analytics.csv")


numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr = numeric_df.corr()

sns.heatmap(corr[['Attrition']].sort_values(by='Attrition', ascending=False), annot=True, cmap="coolwarm")
plt.title("Correlation with Attrition")
plt.show()

```

- Typical finding: Attrition sometimes increases for very low or very high project counts; employees overloaded with many projects (e.g., 6+) show higher attrition risk.

- Practical action: Monitor workload, balance project assignments, offer compensation or recognition to high-load employees.
